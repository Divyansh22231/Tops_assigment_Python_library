```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
np.random.seed(42)

# Q1
delivery_times = np.random.randint(15, 61, 50)

sns.histplot(delivery_times,bins=10)
plt.title("Distribution of Zomato Delivery Times")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Number of Orders")
```

```python
# Q2
df = sns.load_dataset("tips")
df
sns.boxplot(y="total_bill", x="day", data=df)
```

```python
# Q3
mi = np.random.randint(49, 270, 20)
csk = np.random.randint(49, 270, 20)
rcb = np.random.randint(49, 270, 20)
kkr = np.random.randint(49, 270, 20)
srh = np.random.randint(49, 270, 20)
dc = np.random.randint(49, 270, 20)
pbks = np.random.randint(49, 270, 20)
rr = np.random.randint(49, 270, 20)

df = pd.DataFrame({
    'MI': mi,
    'CSK': csk,
    'RCB': rcb,
    'KKR': kkr,
    'SRH': srh,
    'DC': dc,
    'PBKS': pbks,
    'RR': rr
})

sns.violinplot(data=df)
sns.set_style("darkgrid")
```

```python
# Q4
np.random.seed(42)
genres=np.random.choice(["pop","Rock","Hip-Hop","Jazz"],size=40)
df1=pd.DataFrame({
  'Genere':genres
})
sns.countplot(data=df1,x='Genere')
plt.title('Number of Songs per Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Songs')
```

```python
# Q5
steps = [6500, 8200, 7500, 9100, 12000, 10500, 8700]

sns.set_theme(style='whitegrid')

sns.kdeplot(steps, fill=True, color='green')

plt.title('Distribution of Daily Step Counts')
plt.xlabel('Steps')
plt.ylabel('Density')
plt.show()
```
