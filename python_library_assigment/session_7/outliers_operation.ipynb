```python
import pandas as pd
```

```python
zomato_datsett=pd.read_csv(r"C:\Users\Divyansh yadav\Desktop\python_library\zomato_ratings_dataset.csv")

df=pd.DataFrame(zomato_datsett)
df
```

```python
Q1=df['rating'].quantile(0.25)
Q3=df['rating'].quantile(0.75)

IQR=Q3-Q1

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

outliers=df[(df['rating']<lower_bound) | (df['rating']>upper_bound)]

print(outliers.index.tolist())
```
