```python
df1 = pd.DataFrame({
    'post_id': [101, 102, 103, 104, 105, 106],
    'posted_at': [
        '2024-01-01 12:00:00',
        '2024-01-02 15:30:00',
        '2024-01-03 08:45:00',
        '2024-01-04 20:10:00',
        '2024-01-05 05:00:00',
        '2024-01-06 18:25:00'
    ]
})

# Convert to datetime and set timezone as UTC
df1['posted_at'] = pd.to_datetime(df1['posted_at'], utc=True)

# Convert UTC to Asia/Kolkata
df1['posted_at_kolkata'] = df1['posted_at'].dt.tz_convert('Asia/Kolkata')


print(df1['posted_at_kolkata'].head())
```
