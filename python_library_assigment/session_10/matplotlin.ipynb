```python
# Q1
import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,7]
steps=[10000,2000,3000,5000,8000,4000,9000]

plt.plot(days,steps)
plt.savefig("steps_lineplot.png")
```

```python
# Q2

ratings = [4.4, 4.3, 4.6, 4.5, 4.2, 4.0, 4.1, 4.3, 4.4, 4.5]

prices = [1200, 800, 1800, 900, 700, 400, 300, 1000, 1500, 1300]

plt.figure(figsize=(10,6))

plt.scatter(ratings,prices)

plt.xlabel("Zomato ratings")
plt.ylabel("average meal prices")
```

```python
# Q3
platforms = ['Swiggy', 'Zomato', "Domino's"]
orders = [22, 31, 15]

plt.bar(platforms,orders ,label="orders",color=['orange','red','green'])
plt.xlabel('Platform')
plt.ylabel('Number of Orders')
plt.title('Food Delivery Orders in Last Month')
plt.legend()

```

```python
# Q4
durations = [12, 18, 25, 30, 35, 22, 28, 40, 15, 20,
             45, 50, 38, 27, 33, 17, 24, 42, 29, 36]

plt.hist(durations,bins=5,edgecolor='red',color='yellow')

plt.xlabel('Session Duration (minutes)')
plt.ylabel('Frequency')
plt.title('Spotify Listening Session Durations')
```

```python
# Q5
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
screen_time = [45, 60, 50, 75, 80, 95, 70]  # minutes
likes = [20, 35, 25, 40, 50, 60, 45]

fig,ax=plt.subplots(1,2,figsize=(12,5))
ax[0].plot(days,screen_time,marker='o',color="red")
ax[0].set_title('Daily Instagram Screen Time')
ax[0].set_xlabel('Day')
ax[0].set_ylabel('Screen Time (minutes)')

ax[1].bar(days,likes,color='red')
ax[1].set_title("post like per day")
ax[1].set_xlabel("Day")
ax[1].set_ylabel("number of like")
```
