```python
# Q1
import matplotlib.pyplot as plt

days=[1,2,3,4,5,6,7]
steps=[10000,2000,3000,5000,8000,4000,9000]

ratings = [4.4, 4.3, 4.6, 4.5, 4.2, 4.0, 4.1]
prices = [1200, 800, 1800, 900, 700, 400, 300]

platforms = ['Swiggy', 'Zomato', "Domino's","pizza hurt","dadan handi"]
orders = [22,31,15,18,26,]

durations = [12, 18, 25, 30, 35, 22, 28, 40, 15, 20,
             45, 50, 38, 27, 33, 17, 24, 42, 29, 36]

fig,ax=plt.subplots(2,2 ,figsize=(20,16))

# 1st chart
ax[0,0].plot(days,steps,marker='o',color='red')
ax[0,0].set_xlabel("Days",color='green',fontsize='16')
ax[0,0].set_ylabel("Steps",color='green',fontsize='16')

#2nd chart
ax[0,1].scatter(ratings,prices)
ax[0,1].set_xlabel("Zomato ratings",color='red',fontsize='16')
ax[0,1].set_ylabel("average meal prices",color='red',fontsize='16')

#3rd chart
ax[1,0].hist(durations)
ax[1,0].hist(durations,bins=5,edgecolor='red',color='yellow')
ax[1,0].set_xlabel('Session Duration (minutes)',fontsize='16')
ax[1,0].set_ylabel('Frequency',fontsize='16')
ax[1,0].set_title('Spotify Listening Session Durations',fontsize='16')

# 4rth chart
ax[1,1].bar(platforms,orders ,label="orders",color=['orange','red','green'])
ax[1,1].set_xlabel('Platform',fontsize='16',color='blue')
ax[1,1].set_ylabel('Number of Orders',fontsize='16',color='blue')
ax[1,1].set_title('Food Delivery Orders in Last Month',fontsize='16')
ax[1,1].legend()


```

```python
# Q4
movies=["cocktail","rabdaa","kabir singh","virat kohli","mocktail"]

ratings=[4,3.4,5,4.5,6]

plt.scatter(movies,ratings)

for i in range(len(ratings)):
  plt.annotate(
    movies[i],
        xy=(movies[i], ratings[i]),
        xytext=(0, 10),
         textcoords="offset points"
  )
```
