```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Q1
df=sns.load_dataset("tips")
sns.pairplot(data=df,)
```

```python
# Q2

# done with the help of chatgpt
df1=sns.load_dataset("flights")
flights_pivot = df1.pivot_table(
    index="month",
    columns="year",
    values="passengers"
)

plt.figure(figsize=(12, 6))
sns.heatmap(
    flights_pivot,
    annot=True,
    fmt=".0f",
    cmap="YlOrRd"
)

plt.title("Flights Dataset Heatmap (Passengers by Month and Year)")
plt.xlabel("Year")
plt.ylabel("Month")

```
