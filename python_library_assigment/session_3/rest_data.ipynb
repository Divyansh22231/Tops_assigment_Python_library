```python
import numpy as np

# Q1
# Daily step counts of two friends for 7 days
friend1 = np.array([8000, 10000, 7500, 9000, 12000, 11000, 9500])
friend2 = np.array([7000, 9500, 8000, 8500, 10000, 10500, 9000])


addition = friend1 + friend2

subtraction = friend1 - friend2


multiplication = friend1 * friend2


division = friend1 / friend2

print("Friend 1 Steps:", friend1)
print("Friend 2 Steps:", friend2)

print("\nAddition:")
print(addition)

print("\nSubtraction:")
print(subtraction)

print("\nMultiplication:")
print(multiplication)

print("\nDivision:")
print(division)
```

```python
# Q2
friend1=np.array([3,4,5,
                2,4,1,
                7,6,8])

friend2=np.array([4,3,5,
                2,7,1,
                7,6,1])

dot_result = np.dot(friend1, friend2)

# Matrix multiplication
matmul_result = np.matmul(friend1, friend2)

print("Dot Product:", dot_result)
print("Matmul Result:", matmul_result)
```

```python
# Q3
matrix=np.array([3,4,5,4,
                2,4,1,6,
                7,6,8,7,
                4,3,2,1]).reshape((4,4))

transposed_matrix=np.transpose(matrix)
print(transposed_matrix)

print("mean :",np.mean(matrix))
print("standard deviation : ",np.std(matrix))
print("variance : ",np.var(matrix))
```

```python
# Q4
A = np.array([[4, 2],
              [1, 3]])

print("Matrix:",A)


print("Determinant:",np.linalg.det(A))

print("Inverse:",np.linalg.inv(A))


eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:",eigenvalues)

print("Eigenvectors:",eigenvectors)
```
